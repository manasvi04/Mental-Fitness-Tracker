```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
for i in df1.columns:
  if df1[i].dtype=='object':
   df1[i]=l.fit_transform(df1[i])
```

```python
df1.shape
```

```python
x = df1.drop('Mental Fitness',axis=1)
y=df1['Mental Fitness']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=.20,random_state=2)
```

```python
print("x_train :",x_train.shape)
print("x_test :",x_test.shape)
print("\ny_train :",y_train.shape)
print("y_test :",y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
lr=LinearRegression()
lr.fit(x_train,y_train)

y_pred=lr.predict(x_train)

mse=mean_squared_error(y_train,y_pred)

rmse=(np.sqrt(mse))

r2=r2_score(y_train,y_pred)

print("The Linear Regression model performance for training set")
print("<------------------------------------------------------>")
print("MSE is {}",format(mse))
print("RSME is {}",format(rmse))
print("R2 score is {}",format(r2))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(x_train,y_train)

y_pred=rf.predict(x_train)

mse=mean_squared_error(y_train,y_pred)

rmse=(np.sqrt(mse))

r2=r2_score(y_train,y_pred)

print("The Round Forest Regressor model performance for training set")
print("<------------------------------------------------------>")
print("MSE is {}",format(mse))
print("RSME is {}",format(rmse))
print("R2 score is {}",format(r2))
```

```python
y_pred=lr.predict(x_test)

mse=mean_squared_error(y_test,y_pred)

rmse=(np.sqrt(mse))

r2=r2_score(y_test,y_pred)

print("The Linear Regression model performance for training set")
print("<------------------------------------------------------>")
print("MSE is {}",format(mse))
print("RSME is {}",format(rmse))
print("R2 score is {}",format(r2))



y_pred=rf.predict(x_test)

mse=mean_squared_error(y_test,y_pred)

rmse=(np.sqrt(mse))

r2=r2_score(y_test,y_pred)

print("The Round Forest Regressor model performance for training set")
print("<------------------------------------------------------>")
print("MSE is {}",format(mse))
print("RSME is {}",format(rmse))
print("R2 score is {}",format(r2))
```
